```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_random = np.random.rand(100)*100
y_random = np.random.rand(100)*100
```

```python
fig = plt.figure(figsize=[10,8])
ax = fig.add_subplot(111, facecolor='Black',xlim=(0,100), ylim=(0,100))
ax.scatter(x_random,y_random, marker='*', color='w')
plt.grid(True)
plt.show()
```
